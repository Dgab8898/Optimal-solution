```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import optimize
```

```python
# Define the objective function that wll minimw3:

def func(x):
    return ((x[0] - 0.5) **2 + (x[1] + 0.5)**2) * np.cos(0.5*x[0] *x[1])
```

```python
# Next create a grid of values that we can plot our objective function

x_r = np.linspace(-1, 1)
y_r = np.linspace(-2, 2)
x, y = np.meshgrid(x_r, y_r)
```

```python
# Now we can evaluate the function on this grid of points:

z = func([x, y])
```

```python
# Next we create  a new figure with 3d axes object and set the a title

fig = plt.figure(tight_layout=True)
ax = fig.add_subplot(projection="3d")
ax.tick_params(axis="both", which="major", labelsize=9)
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.set_title("Objective function")


#Now we can plot the objective function as a surface on the axes we just created:
ax.plot_surface(x, y, z, alpha=0.7)


# We chooose an initial point that our minimix=zation routine will start its iteration at and plot this on the surface:

x0 =np.array([-0.5, 1.0])
ax.plot(x0[[0]], [x0[1]], func(x0), "r")
```

```python
# Then we used minimize routine from the optimize package to find the minmimum value and print the result
result = optimize.minimize(func,x0, tol=1e-6, method= "Nelder-Mead")
print(result)
```

```python
# Finally we plot the minmum value found by the minmize routine on the top of the objective function surface
ax.plot([result.x[0]], [result.x[1]], [result.fun], "r*")
```
